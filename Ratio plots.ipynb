```python
def ratio_plot(h1,h2, key):
    hist1, hist2 = h1[key], h2[key]
    x = [b.xMid for b in hist1.bins]
    ratio = [b1.height/b2.height/hist1.integral()*hist2.integral() 
              for b1,b2 in zip(hist1.bins, hist2.bins)]
    plt.plot(x, ratio, linewidth=3)
    
```

```python
key =  '/VBF/WW_pT'
ratio_plot(histoswpwpQCD, histoswpwpEW, key=key)
plt.ylabel(r"$W^+W^+\frac{QCD}{EW}$")
plt.ylim(0,2)
plt.xlim(0,1000)
plt.xlabel("$p_T (WW)$ (Gev)")
plt.savefig("ppqcd_ppew_wwpt.pdf")
```

```python
key =  '/VBF/WW_pT'
ratio_plot(histoswpwpEW, histoswmwmEW, key=key)
plt.ylim(0,2)
plt.xlim(0,400)
plt.ylabel(r"$\frac{W^+W^+}{W^-W ^-}[EW]$")
plt.xlabel("$p_T (WW)$ (Gev)")
plt.savefig("ppew_mmew_wwpt.pdf")
```
